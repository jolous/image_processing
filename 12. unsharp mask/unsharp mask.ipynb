```python
from skimage import io, img_as_float
from skimage.filters import unsharp_mask
from skimage.filters import gaussian
from matplotlib import pyplot as plt
```

```python
# unsharp image without unsharp_mask function
# unsharpened image = original + amount * (original - blurred)
img = img_as_float(io.imread('../sample/1.png', as_gray=True))
gaussian_img = gaussian(img, sigma=2, mode='constant', cval=0.0)
img2 = (img - gaussian_img) * 2
img3 = img + img2

plt.imshow(img3, cmap="gray")
```

```python
plt.figure(figsize=(12,6))

plt.subplot(121) # 1 row 2 columns
plt.title('original pic')
plt.imshow(img, cmap="gray")

plt.subplot(122)
plt.title('unsharp pic')
plt.imshow(img3, cmap="gray")
```

```python
img = img_as_float(io.imread('../sample/1.png', as_gray=True))

unsharped_img = unsharp_mask(img, radius=3, amount=2)

plt.figure(figsize=(12,6))

plt.subplot(121) # 1 row 2 columns
plt.title('original pic')
plt.imshow(img, cmap="gray")

plt.subplot(122)
plt.title('unsharp pic')
plt.imshow(unsharped_img, cmap="gray")
```

```python
# good for images with lots of particles 
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('../sample/1.tiff', 1)

lab_img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab_img)

clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
clahe_img = clahe.apply(l)

CLAHE_img = cv2.merge((clahe_img, a, b))
corrected_image = cv2.cvtColor(CLAHE_img, cv2.COLOR_LAB2BGR)

plt.figure(figsize=(12,8))

plt.subplot(121) 
plt.title('Original image')
plt.imshow(img)

plt.subplot(122)
plt.title('Corrected image')
plt.imshow(corrected_image)
```

```python
from cv2_rolling_ball import subtract_background_rolling_ball

img = cv2.imread('../sample/1.tiff', 0)

radius = 50

final_img, background = subtract_background_rolling_ball(img,
                                                         radius,
                                                         light_background=False,
                                                         use_paraboloid=False,
                                                         do_presmooth=True)

clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
clahe_img = clahe.apply(final_img)

plt.figure(figsize=(12,8))

plt.subplot(121) 
plt.title('Original image')
plt.imshow(img)

plt.subplot(122)
plt.title('After CLAHE')
plt.imshow(final_img)
```

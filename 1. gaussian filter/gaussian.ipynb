```python
from skimage import io, filters
from matplotlib import pyplot as plt
```

```python
img = io.imread('../sample/1.png')
gaussian_img = filters.gaussian(img, sigma=1)

plt.imshow(gaussian_img)
```

```python
def gaussian_of_img(img, sigma=1):
    gaussian_img = filters.gaussian(img, sigma)
    return(gaussian_img)
```

```python
my_img = io.imread('../sample/1.png')
filtered = gaussian_of_img(my_img, 3)
plt.imshow(filtered)
```

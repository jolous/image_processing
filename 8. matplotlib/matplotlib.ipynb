```python
from matplotlib import pyplot as plt
%matplotlib inline

x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
```

```python
import numpy as np

a = np.array(x)
b = np.array(y)
plt.plot(a,b)
```

```python
plt.plot(a, b, 'r--') # red dash, 'bo' -> blue dots
```

```python
plt.plot(a, b, 'r--', linewidth=3)
plt.axis([0,6,0,50]) # define range for x and y axies
```

```python
wells = ['well1', 'well2', 'well3', 'well4', 'well5']
cells = [80, 62, 88, 110, 90]

plt.bar(wells, cells)
```

```python
plt.scatter(wells, cells)
```

```python
plt.plot(wells, cells)
```

```python
plt.bar(wells, cells)
plt.scatter(wells, cells)
plt.plot(wells, cells)
```

```python
plt.plot(a,b)

# use setp() to define multiple parameters
fig = plt.plot(a,b)
plt.setp(fig, color='r', linewidth=4.0)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.bar(wells,cells)
plt.xlabel("well #")
plt.ylabel("# dead cells")
plt.title("Dead cells in each well", fontsize=16, color="red")
plt.axis([1, 6, 60, 120]) #xmin, xmax, ymin, ymax
plt.grid(True)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,125,1350,11250,100500]

plt.figure(figsize=(12,6))

#linear
plt.subplot(121) # 1 row 2 columns
plt.plot(x,y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

#log
plt.subplot(122)
plt.plot(x,y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
```

```python
plt.figure(figsize=(16,6))

#1 row 3 columns
plt.subplot(131)
plt.bar(wells,cells)

plt.subplot(132)
plt.scatter(wells,cells)

plt.subplot(133)
plt.plot(wells,cells)

plt.suptitle("Multiple Plots")
plt.show()
```

```python
plt.figure(figsize=(10,10))

#2 rows 2 columns
plt.subplot(221)
plt.bar(wells,cells)

plt.subplot(222)
plt.scatter(wells,cells)

plt.subplot(223)
plt.plot(wells,cells)

plt.suptitle("Multiple Plots")
plt.show()

```

```python
fig = plt.figure(figsize=(16,6))

ax1 = fig.add_subplot(131)
ax1.set(title='vertical bar', xlabel='well #', ylabel='w')

ax2 = fig.add_subplot(132)
ax2.set(title='horizontal bar', xlabel='well #', ylabel='w')

ax3 = fig.add_subplot(133)

#plot the data
ax1.bar(wells, cells)
ax1.barh(wells, cells)
ax1.plot(wells, cells)

plt.show()
```

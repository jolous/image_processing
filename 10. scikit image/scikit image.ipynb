```python
from matplotlib import pyplot as plt

from skimage import io, color
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
img = io.imread('../sample/1.tiff', as_gray=True)
img.shape
```

```python
img_rescaled = rescale(img, 0.25, anti_aliasing=False) # rescale by 25%
img_rescaled.shape
```

```python
img_resized = resize(img, (200, 200), anti_aliasing=True) # image size 200 * 200  
img_resized.shape
```

```python
img_downscaled = downscale_local_mean(img, (4, 3))
```

```python
plt.imshow(img_downscaled, cmap='gray')
```

```python
from skimage.filters import gaussian, sobel

img = io.imread('../sample/1.tiff')
plt.imshow(img)
```

```python
gaussian_using_skimage = gaussian(img, sigma=1, mode='constant', cval=0.0)
plt.imshow(gaussian_using_skimage)
```

```python
img_gray = io.imread('../sample/1.tiff', as_gray=True)
sobel_image = sobel(img_gray)
plt.imshow(sobel_image, cmap='gray')
```

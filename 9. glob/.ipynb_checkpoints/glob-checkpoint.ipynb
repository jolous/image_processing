```python
import cv2
import glob
```

```python
file_list = glob.glob('../sample/*.*')
print(file_list)
```

```python
my_list=[]

path = "../sample/*.*"
for file in glob.glob(path):
    print(file)
    a = cv2.imread(file)
    my_list.append(a)
    
from matplotlib import pyplot as plt
plt.imshow(my_list[1])
```

```python
import os

path = "../sample/"
print(os.listdir(path))
print('****')
for image in os.listdir(path):
    print(image)
```

```python
import os

for root, dirs, files in os.walk('../sample/'):
    path = root.split(os.sep)
    print((len(path) - 1) * '---', os.path.basename(root))
    for file in files:
        print(len(path) * '---', file)
```

```python
import os

for root, dirs, files in os.walk('../sample/'):
    for name in dirs:
        print(os.path.join(root, name))
    for name in files:
        print(os.path.join(root, name))
```

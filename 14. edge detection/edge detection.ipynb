```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('../sample/1.tiff', 0)

# Edge detection
from skimage.filters import roberts, sobel, scharr, prewitt

roberts_img = roberts(img)
sobel_img = sobel(img) # best one
scharr_img = scharr(img)
prewitt_img = prewitt(img)

plt.figure(figsize=(12,8))

plt.subplot(221) 
plt.title('roberts')
plt.imshow(roberts_img, cmap='gray')

plt.subplot(222)
plt.title('sobel')
plt.imshow(sobel_img, cmap='gray')

plt.subplot(223)
plt.title('scharr')
plt.imshow(scharr_img, cmap='gray')

plt.subplot(224)
plt.title('prewitt')
plt.imshow(prewitt_img, cmap='gray')
```

```python
# Canny edge detection

canny_edge = cv2.Canny(img, 50, 80)

# Autocanny
sigma = 0.3
median = np.median(img)

# apply automatic Canny edge detection using the computed median
# lower threshold is sigma % lower than median
# if the value is below 0 then take 0 as the value
lower = int(max(0, (1.0 - sigma) * median))

# upper threshold is sigma % higher than median
# if the value is larger than 255 then take 255 as the value
upper = int(min(255, (1.0 - sigma) * median))

auto_canny = cv2.Canny(img, lower, upper)

plt.figure(figsize=(12,8))


plt.subplot(121)
plt.title('canny_edge')
plt.imshow(canny_edge, cmap='gray')

plt.subplot(122)
plt.title('auto_canny')
plt.imshow(auto_canny, cmap='gray')
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('../sample/1.tiff', 1) # color is BRG not RGB
print(img.shape)
# resize two times bigger
resized = cv2.resize(img, None, fx=2, fy=2, interpolation= cv2.INTER_CUBIC)
```

```python
plt.figure(figsize=(12,6))

plt.subplot(121) # 1 row 2 columns
plt.title('original pic')
plt.imshow(img)

plt.subplot(122)
plt.title('resized pic')
plt.imshow(resized)
```

```python
blue = img[:,:,0] # show only blue  (BGR so B = 0)
green = img[:,:,1] # green
red = img[:,:,2] # red only
```

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.title('blue')
plt.imshow(blue)

plt.subplot(132)
plt.title('green')
plt.imshow(green)

plt.subplot(133)
plt.title('red')
plt.imshow(red)
```

```python
b, g, r = cv2.split(img)

plt.figure(figsize=(12,6))

plt.subplot(131)
plt.title('blue')
plt.imshow(b)

plt.subplot(132)
plt.title('green')
plt.imshow(g)

plt.subplot(133)
plt.title('red')
plt.imshow(r)
```

```python
img_merged = cv2.merge((b,g,r))
plt.imshow(img_merged)
```

```python
# edge detection
img = cv2.imread('../sample/1.tiff')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 100, 200) # Image, min and max

plt.figure(figsize=(12,6))

plt.subplot(121)
plt.title('gray')
plt.imshow(gray, cmap='gray')

plt.subplot(122)
plt.title('edge')
plt.imshow(edges, cmap='gray')
```

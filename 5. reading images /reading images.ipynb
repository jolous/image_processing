```python
from skimage import io, img_as_float, img_as_ubyte

img = io.imread('../sample/1.png')
print(img.shape) # y, x , c - c is number of channels

img2 = img_as_float(img) # convert the numbers between 0 and 1
print(img2.shape)
```

```python
print(f"img: {img[0][0]}")
print(f"img2: {img2[0][0]}")
```

```python
img3 = img_as_ubyte(img2) # conver the float to the original format
print(f"img3: {img3[0][0]}")
```

```python
import cv2
gray_img = cv2.imread('../sample/1.png', 0) # you can read as a gray image if you put 0
print(gray_img.shape)
print(f"gray_img: {gray_img[0][0]}")
```

```python
cv_img = cv2.imread('../sample/1.png') # BGR instead of RGB
print(cv_img.shape)
print(f"cv_img: {cv_img[0][0]}")
```
